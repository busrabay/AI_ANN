```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

LR = LogisticRegression()
Svc = SVC()
RF = RandomForestClassifier()
DT = DecisionTreeClassifier()
KNN = KNeighborsClassifier(n_neighbors=25)

LR.fit(X_train, y_train)
Svc.fit(X_train, y_train)
RF.fit(X_train, y_train)
DT.fit(X_train, y_train)
KNN.fit(X_train, y_train)


print(f"LR accuracy: {LR.score(X_test, y_test)}")
print(f"SVC accuracy: {Svc.score(X_test, y_test)}")
print(f"RF accuracy: {RF.score(X_test, y_test)}")
print(f"DT accuracy: {DT.score(X_test, y_test)}")
print(f"KNN accuracy: {KNN.score(X_test, y_test)}")
```

```python
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential, load_model

giris_katmani = X_train.shape[1]
print(f"input_dim: {giris_katmani}")

output_dim = y_train
print(f"output_dim: {output_dim}")

model = Sequential()
model.add(Dense(8, input_dim=giris_katmani, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary() # modelin özeti

model.fit(X_train, y_train, epochs=5, batch_size=10, validation_data=(X_test, y_test))
```

```python
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential, load_model
giris_katmani = X_train.shape[1]
print(f"input_dim: {giris_katmani}")

output_dim = y_train
print(f"output_dim: {output_dim}")

model = Sequential()
# alternatif aktivasyon fonksiyonları: relu, tanh, sigmoid, softmax
model.add(Dense(128, input_dim=giris_katmani, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# alternatif optimizasyon fonksiyonları: adam, rmsprop, sgd, adagrad, adadelta, adamax, nadam
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.summary() # modelin özeti

model.fit(X_train, y_train, epochs=3, batch_size=25, validation_data=(X_test, y_test))
```
